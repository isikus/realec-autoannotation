```python
lic = pd.DataFrame({
    'entry': [],
    'substring': []
})
'''
for r in dummyrows:
    fpath = r[0]
    start = r[1]
    end = r[2]
'''
for index, row in lex_item_choice.iterrows():
    if row['string_tokens'] != 1:
        continue
    fpath = row['file']
    start = row['start']
    end = row['end']
    with open(fpath[:-3]+'txt', 'r', encoding='utf-8') as f:
        raw = f.read()
        substr = raw[start:end]
        raw = raw[:start] + '[MASK]' + raw[end:]
        raw = re.sub(r'\s', ' ', raw)
        raw = re.sub(r'( )+', ' ', raw)
        sentences = nltk.sent_tokenize(raw)
        for i in range(len(sentences)):
            if '[MASK]' in sentences[i]:
                if i < 2:
                    lic.loc[len(lic)] = [' '.join(sentences[0:i+3]), substr]
                else:
                    lic.loc[len(lic)] = [' '.join(sentences[i-2:i+3]), substr]
                break
lic
```

# Контексты без ошибки

```python
import os

TxtPaths = []

for root, dirs, files in os.walk('.'):
    for file in os.listdir(root):
        if file.endswith(".txt"):
            path = root+os.sep+file
            TxtPaths.append(path)
```

```python
TxtPaths
```

```python
import random

non_lic = pd.DataFrame({
    'entry': [],
    'substring': []
})

from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()

from nltk.tokenize.moses import MosesDetokenizer
detokenizer = MosesDetokenizer()


contexts_needed = len(lic) * 15
Min = min([len(nltk.word_tokenize(entry)) for entry in lic['entry'].values])
# contexts_needed = 5
I = 0
while I < contexts_needed:
    fpath = random.choice(TxtPaths)
    with open(fpath, 'r', encoding='utf-8') as f:
        raw = f.read()
        raw = re.sub(r'\s', ' ', raw)
        raw = re.sub(r'( )+', ' ', raw)
        wordsoup = tknzr.tokenize(raw)
        if len(wordsoup) < Min:
            continue
        while True:
            seed = random.randint(0, len(wordsoup)-1)
            if not re.search(r'[a-zA-Z]', wordsoup[seed]):
                continue
            else:
                substr = wordsoup[seed]
                wordsoup[seed] = '[MASK]'
                break
        masked_text = detokenizer.detokenize(wordsoup, return_str=True)
        
        sentences = nltk.sent_tokenize(masked_text)
        for i in range(len(sentences)):
            if '[MASK]' in sentences[i]:
                if i < 2:
                    entry = ' '.join(sentences[0:i+3])
                else:
                    entry = ' '.join(sentences[i-2:i+3])
                break
    if entry in lic['entry'].values:
        continue
    elif entry in non_lic['entry'].values:
        continue
    else:
        non_lic.loc[len(non_lic)] = [entry, substr]
        if not (I % 100):
            print(str(I)+'/'+str(contexts_needed),'entries written ('+'{:.2f}'.format((I/contexts_needed)*100)+'%)')
        I += 1

non_lic
```
